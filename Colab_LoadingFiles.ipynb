# Loading Files in Colab
There are several ways to load files into a Colab notebook. We will cover two of the most common.

1. From your Google Drive
2. From a Local Directory

## Loading Files from Your Google Drive
1. First, download and transfer the file you want to work with into a folder on your Google Drive
2. Then run the following code in a code cell
```
from google.colab import drive
drive.mount('/content/drive')
```
>- After running this in a code cell you will get a prompt to authorize Colab to access your google drive.
3. Then use the Colab drive navigation to find your file
4. Click the 3 dots next to your file name, then click 'Copy path'
5. Paste the path to your file as a string with an appropriate variable name
6. Read in the file using `pd.read_csv(nameOfFilePath)` or `pd.read_excel(nameOfFilePath)`
